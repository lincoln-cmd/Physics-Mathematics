```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

# f(x, y) = (2x^2 + y^2) x exp(-(x^2 + y^2))

### 함수 정의 및 행렬, 2차원 그래프

```python
def f(x, y):
    r = 2 * (x ** 2) + y**2 # (2x^2 + y^2)
    answer = r * np.exp(-r)
    return answer
```

```python
# x, y 범위 설정 및 9등분
xn = 9
x = np.linspace(-2, 2, xn)
y = np.linspace(-2, 2, xn)
print(x,y)
```

```python
temp = np.zeros((len(x), len(y))) # 2차원 배열
for i in range(xn):
    for j in range(xn):
        temp[j, i] = f(x[i],y[j]) # (0,0), (1,0), (2,0) ... (0,1), (1,1), (2,1) 순서로 저장
```

```python
print(np.round(temp, 1))
```

```python
# 2차원 배열의 요소를 색으로 표현

plt.figure(figsize = (3.5,3)) # 크기 설정
plt.gray()
plt.pcolor(temp) # plt.pcolor(2차원 narray), 행렬을 색상으로 표현
plt.colorbar()
plt.show()
```

### 3차원 그래프로 그리기

```python
xx, yy = np.meshgrid(x, y) # 좌표점 x, y로 새로운 좌표점 xx, yy 생성
print(xx, yy)
```

```python
plt.figure(figsize = (5, 3.5))
ax = plt.subplot(1, 1, 1, projection = '3d') # 3차원
ax.plot_surface(xx, yy, temp, rstride = 1, cstride = 1, alpha = 0.3,
               color = 'blue', edgecolor = 'black') # 함수의 표면 표시
plt.show()

# cstride, rstride : 가로, 세로 선 갯수 지정
# alpha : 투명도. 0 ~ 1 사이의 실수, 0에 가까울수록 투명
```

```python
plt.figure(figsize = (5, 3.5))
ax = plt.subplot(1, 1, 1, projection = '3d')
ax.plot_surface(xx, yy, temp, rstride = 1, cstride = 1, color = 'blue', alpha = 0.2,
               edgecolor = 'red')
ax.view_init(75, -95)
    # view_init(a, b) : 3차원 그래프의 방향 조정.
    # a -> 상하회전각(0 : 옆에서 본 각도, 90 : 위에서 본 각도)
    # b -> 좌우회전각(양수 : 시계방향으로 회전, 음수 : 반시계방향으로 회전)
ax.set_zticks((0, 0.2))
plt.show()
```

```python
xn = 50
x = np.linspace(-2, 2, xn)
y = np.linspace(-2, 2, xn)
temp = np.zeros((len(x), len(y)))
for i in range(xn):
    for j in range(xn):
        temp[j, i] = f(x[i], y[j])
plt.figure(figsize = (3.5, 3))
plt.gray()
plt.pcolor(temp)
plt.colorbar()
plt.show()

xx, yy = np.meshgrid(x, y)
plt.figure(figsize = (5, 3.5))
ax = plt.subplot(1, 1, 1, projection = '3d')
ax.plot_surface(xx, yy, temp, rstride = 1, cstride = 1, color = 'blue', alpha = 0.4, edgecolor = 'black')
ax.set_zticks((0, 0.2))
ax.view_init(70, 90)
plt.show()
```

### 등고선 그리기

```python
plt.figure(figsize = (4, 4))
cont = plt.contour(xx, yy, temp, 5, colors = 'black') # 등고선의 높이를 5단계로 표시
cont.clabel(fmt = '%3.2f', fontsize = 8) # 등고선의 높이, 표시할 형식 설정
plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$y$', fontsize = 14)
plt.show()
```
